```python
import numpy as np
import pandas as pd
import folium
from folium.plugins import MarkerCluster
```

```python
df = pd.read_csv("significant-earthquake-database.csv", sep = ";")
```

```python
map = folium.Map()
```

```python
liste_pays = df['Country'].unique()
for pays in liste_pays:
    cluster = MarkerCluster().add_to(map)
    for i in range(len(df[df['Country']==pays].dropna())):
        element = df[df['Country']==pays]["Coordinates"].dropna().iloc[i]
        annee = df[df['Country']==pays]["Year"].dropna().iloc[i]
        intensite = df[df['Country']==pays]["EQ Primary"].dropna().iloc[i]
        coordonnees = element.split(',')
        coordonnees[0], coordonnees[1] = float(coordonnees[0]), float(coordonnees[1])
        folium.Marker(coordonnees, popup = f"anne : {annee}, intensit : {intensite}").add_to(cluster)
map
```
